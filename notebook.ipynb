```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd

data = pd.read_csv("datasets/office_episodes.csv")

print(data.head())
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize =(12,8))

plt.scatter(data['episode_number'], data['viewership_mil'])

plt.title("Popularity, Quality, and Guest Appearances on the Office")

plt.xlabel("Episode Number")

plt.ylabel("Viewership (Millions)")

plt.show()
```

```python
# Define an empty list

colors = []

for lab, row in data.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append("red")
    
    elif row['scaled_ratings'] >= 0.25 and row['scaled_ratings'] < 0.50:
        colors.append("orange") 
    
    elif row['scaled_ratings'] >= 0.50 and row['scaled_ratings'] < 0.75:
        colors.append("lightgreen")
    
    else :
        colors.append("darkgreen")
        
colors[:10]    
```

```python
sizes = []

for lab, row in data.iterrows():
    if row['has_guests'] == "False":
        sizes.append(25)
    else:
        sizes.append(250)
sizes[:10] 
```

```python
data['colors'] = colors
data['sizes'] = sizes

data.info()
non_guest_df = data[data['has_guests'] == False]
guest_df = data[data['has_guests'] == True]
       
```

```python

import matplotlib.pyplot as plt

fig = plt.figure()

plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')

plt.scatter(x= non_guest_df['episode_number'], 
            y= non_guest_df['viewership_mil'], c=non_guest_df['colors'], s=25)

plt.scatter(x= guest_df['episode_number'], 
            y= guest_df['viewership_mil'], c = guest_df['colors'], s=250,
            marker ="*"
           )

plt.title("Popularity, Quality, and Guest Appearances on the Office")

plt.xlabel("Episode Number")

plt.ylabel("Viewership (Millions)")

plt.show()

```
